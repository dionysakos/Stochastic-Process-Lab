```python
import simple_markov_chain_lib as lib
markov_table = {
    'Κ': {'Β': .5, 'Σ': .5},
    'Β': {'Κ': .5, 'Σ': .5},
    'Σ': {'Κ': .25, 'Β': .25, 'Υ': .25, 'Μ': .25},
    'Υ': {'Σ': 1.},
    'Μ': {'Σ': 1.}
}

# Αρχική κατανομή:
init_dist = {'Μ': 1.}

mc = lib.markov_chain(markov_table, init_dist)
mc.start()
print("Αρχικά, το έντομο βρίσκεται στο δωμάτιο:", mc.running_state)
for i in range(1, 21):
    mc.move()
    print(f"Βήμα {mc.steps}: Το έντομο βρίσκεται στο δωμάτιο:", mc.running_state)
```
