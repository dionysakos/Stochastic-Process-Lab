```python
import random
import statistics as stat
from simple_markov_chain_lib import markov_chain

random.seed(2018)
p = 1/6
init_probs = {1: 1.0}
markov_table = {
    1: {2: 1.},
    2: {2: 2/3, 3: 1/3},
    3: {1: p, 2: 1-p}
}

mc = markov_chain(markov_table, init_probs)

M = 50       # Πλήθος εκτιμήσεων
N = 200      # Πλήθος δειγμάτων ανά εκτίμηση
steps = 20

estimates = []

for k in range(M):
    counter = 0
    for i in range(N):
        mc.start()
        for j in range(steps):
            mc.move()
        if mc.running_state == 1:
            counter += 1

    phat = counter / N
    estimates.append(phat)

mean_est = stat.mean(estimates)
var_est = stat.variance(estimates)

print(f"--- Αποτελέσματα για N = {N} ---")
print(f"Δειγματική Μέση Τιμή (Mean): {mean_est:.5f}")
print(f"Δειγματική Διασπορά (Variance): {var_est:.5f}")
```
